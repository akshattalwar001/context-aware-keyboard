```python
import requests

urls = [
    "https://www.gutenberg.org/files/1342/1342-0.txt",  # Pride and Prejudice
    "https://www.gutenberg.org/files/11/11-0.txt",       # Alice in Wonderland
    "https://www.gutenberg.org/files/1661/1661-0.txt",   # Sherlock Holmes
    "https://www.gutenberg.org/files/98/98-0.txt",       # A Tale of Two Cities
    "https://www.gutenberg.org/files/84/84-0.txt",       # Frankenstein
    "https://www.gutenberg.org/files/1952/1952-0.txt",   # The Yellow Wallpaper
    "https://www.gutenberg.org/files/174/174-0.txt",     # Dorian Gray
]

combined = ""

for url in urls:
    response = requests.get(url)
    response.encoding = "utf-8"
    combined += response.text + "\n"
    print(f"Downloaded: {url.split('/')[-1]}")

# Save to file
with open("/content/books.txt", "w", encoding="utf-8") as f:
    f.write(combined)

print("Done! Total characters:", len(combined))
```

```python
def clean_gutenberg(text):
    # Remove Gutenberg header and footer
    text = re.sub(r'\*\*\* START OF.*?\*\*\*', '', text, flags=re.DOTALL)
    text = re.sub(r'\*\*\* END OF.*?\*\*\*', '', text, flags=re.DOTALL)

    # Remove [Illustration: ...] blocks
    text = re.sub(r'\[Illustration[^\]]*\]', '', text, flags=re.DOTALL)

    # Remove italic underscores
    text = text.replace('_', '')

    # Remove chapter headings (all caps lines)
    text = re.sub(r'\n[A-Z\s]{5,}\n', '\n', text)

    # Remove extra whitespace and blank lines
    text = re.sub(r'\n{2,}', '\n', text)
    text = re.sub(r'[ \t]+', ' ', text)
    text = text.strip()

    return text

# Clean the text
combined = clean_gutenberg(combined)

# Skip everything before "PREFACE" or "Chapter" — real content starts there
# Find the first occurrence of a real sentence pattern
match = re.search(r'\bChapter\b|\bCHAPTER\b|\bPREFACE\b', combined)
if match:
    combined = combined[match.start():]

print(combined[:500])
print("\nTotal characters:", len(combined))
```

```python
# Fix smart quotes and dashes
combined = combined.replace('"', '"').replace('"', '"')
combined = combined.replace(''', "'").replace(''', "'")
combined = combined.replace('—', '-')

# Remove BOM and Windows line endings
combined = combined.replace('\ufeff', '')
combined = combined.replace('\r', '')

# Remove leftover symbols and brackets
combined = re.sub(r'[£½™•&$%#\[\]]', '', combined)

# Final whitespace cleanup
combined = re.sub(r'[ \t]+', ' ', combined)
combined = re.sub(r'\n{2,}', '\n', combined)
combined = combined.strip()

# Verify
odd_chars = [c for c in set(combined) if not c.isalpha() and not c in " .,!?;:'\"-\n()0123456789"]
print("Remaining unusual characters:", odd_chars)
print("\nSample:", combined[:500])
print("\nTotal characters:", len(combined))
```

```python
# Force-replace all remaining quote variants using unicode codes
combined = combined.replace('\u201c', '"').replace('\u201d', '"')  # " "
combined = combined.replace('\u2018', "'").replace('\u2019', "'")  # ' '
combined = combined.replace('\u0060', "'")  # backtick

# Remove asterisks and slashes
combined = combined.replace('*', '').replace('/', ' ')

# Final check
odd_chars = [c for c in set(combined) if not c.isalpha() and not c in " .,!?;:'\"-\n()0123456789"]
print("Remaining unusual characters:", odd_chars)
print("Total characters:", len(combined))
```
